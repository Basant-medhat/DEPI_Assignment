```python
import pandas as pd


df = pd.read_csv("/content/Titanic.csv")

print(df.info())
df.head()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set visualization style
sns.set_style("whitegrid")

# Plot survival count
plt.figure(figsize=(6,4))
sns.countplot(data=df, x="Survived", palette="coolwarm")
plt.xticks(ticks=[0,1], labels=["Did Not Survive", "Survived"])
plt.xlabel("Survival")
plt.ylabel("Count")
plt.title("Survival Count in Titanic Dataset")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x="Sex", hue="Survived", palette="coolwarm")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.title("Survival Based on Gender")
plt.legend(["Did Not Survive", "Survived"])
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x="Pclass", hue="Survived", palette="coolwarm")
plt.xlabel("Passenger Class")
plt.ylabel("Count")
plt.title("Survival Based on Passenger Class")
plt.legend(["Did Not Survive", "Survived"])
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Age"], bins=30, kde=True, color="blue")
plt.xlabel("Age")
plt.ylabel("Count")
plt.title("Age Distribution of Passengers")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df[df["Survived"]==1]["Age"], bins=30, kde=True, color="green", label="Survived")
sns.histplot(df[df["Survived"]==0]["Age"], bins=30, kde=True, color="red", label="Did Not Survive")
plt.xlabel("Age")
plt.ylabel("Count")
plt.title("Age Distribution by Survival")

plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Fare"], bins=30, kde=True, color="purple")
plt.xlabel("Fare")
plt.ylabel("Count")
plt.title("Fare Distribution of Passengers")
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.boxplot(x="Survived", y="Age", data=df, palette="coolwarm")
plt.xticks(ticks=[0,1], labels=["Did Not Survive", "Survived"])
plt.xlabel("Survival")
plt.ylabel("Age")
plt.title("Age Distribution by Survival")
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.boxplot(x="Pclass", y="Fare", data=df, palette="coolwarm")
plt.xlabel("Passenger Class")
plt.ylabel("Fare")
plt.title("Fare Distribution by Passenger Class")
plt.show()

```

```python
plt.figure(figsize=(8,5))
sns.heatmap(df.isnull(), cmap="viridis", cbar=False, yticklabels=False)
plt.title("Missing Values in Titanic Dataset")
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.violinplot(x="Pclass", y="Age", hue="Survived", data=df, split=True, palette="coolwarm")
plt.xlabel("Passenger Class")
plt.ylabel("Age")
plt.title("Age Distribution by Class and Survival")
plt.show()
```
